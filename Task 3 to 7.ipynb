# List-adding nos from 20 to 30 & removing even nos from the list

```python
numberList=[]
for i in range(20,31):
    numberList.append(i)
print("List of number\n",numberList)
for j in numberList:
    if j%2==0:
        numberList.remove(j)
print("List after removing even nos\n",numberList)
```
